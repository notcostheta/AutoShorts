```python
# Load config Variables
import os
from dotenv import load_dotenv
load_dotenv("config.env")
hf_token = os.getenv("HF_READ_TOKEN")
video_url = "https://www.youtube.com/watch?v=XNQhDl4a9Ko"
```

```python
from app.downloader import download_video, get_audio
video_path = download_video(video_url)
audio_path = get_audio(video_path)
```

```python
from app.transcripts import transcribe_audio

transcript = transcribe_audio(audio_path, hf_token)
```

```python
from app.transform import external_transform
from semantic_router.encoders import HuggingFaceEncoder

transformed_transcript = external_transform(
    transcript,
    encoder=HuggingFaceEncoder(name="Snowflake/snowflake-arctic-embed-l"),
    max_duration=50,
    min_split_tokens=150,
    max_split_tokens=250,
    window_size=5,
    batch_size=32,
)
```

```python
from app.split import make_splits

data = transformed_transcript
video_path = video_path

split_paths = make_splits(data, video_path, output_dir="splits")
```

```python
from app.subtitles import fix_timings, create_ass_from_dfs

data = transformed_transcript

split_dfs = fix_timings(data)
create_ass_from_dfs(dfs=split_dfs, output_folder="splits", max_words=1)
```
